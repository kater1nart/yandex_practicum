### Шаг 2.1 Заполнение пропусков

```python
data[data['total_income'].isna()]
```

```python
21525 - 19351
```

```python
2174/21525 * 100
```

Помимо столбца 'days_employed' пропуски присутствуют в столбце 'total_income', получается у нас нет данных по общему доходу у 2174 человек, что составляет 10.1% от всех данных в каждом из двух столбцов. Почему могли появиться эти пропуки?
Пропуски могли появиться по техническим причинам, человеческий фактор, так же так как пропуски в каждом из двух столбцах находяться в одинаковых строчках, то можно увидить какую нибудь взаимосвязь, может во время опроса если ты не указываешь дней на работе, то и общий доход не имеет значения. Так же есть шанс что некоторые люди в целях конфидециальности намерренно пропускали и не заполняли эти данные, так же есть предположение что для пенсионеров/студентов/безработных эти данные не должны указываться. Поэтому заполним эти пропуски медианными значениями, потому что они наногокорректней подойдут, чем то же среднее значение, потому что могут присутствовать выбросы, которые испортят среднее значение.

```python
total_income_median = data['total_income'].median()
total_income_median
```

```python
data['total_income'] = data['total_income'].fillna(total_income_median)
data.info()
```

```python
data.info()
```
